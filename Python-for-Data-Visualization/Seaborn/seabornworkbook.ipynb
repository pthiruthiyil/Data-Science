```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.distplot(tips['total_bill'],bins=100)
```

```python
sns.distplot(tips['total_bill'],kde=False) #kde = kernal density information. Sum of normal distribution curves against each value in a rug plot
```

```python
sns.kdeplot(tips['total_bill'],shade=True)
```

```python
sns.rugplot(tips['total_bill']) # instead of bins, it will be individual lines.. the x will match with distplot
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips)
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind='hex')
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips, kind='reg')
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind='kde')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
import seaborn as sns

tips = sns.load_dataset("tips")
sns.barplot(x="sex",y="total_bill",data=tips) # By default it will be the mean
```

```python
import numpy as np
sns.barplot(x="sex",y="total_bill",data=tips,estimator=np.std) # std of totalbill .. Any aggregator function can be used
```

```python
sns.countplot(x="sex",data=tips) # No need of y since this is a count plot
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker') # box plot of total_bill per day and per smoker
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips) #by default jitter = True
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips, hue='sex',dodge=True)# dodge is same as split
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)# Similar to strip plot except that points are not overlapped. Does not cale well
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='box') #box,swarm etc.. all others mentioned above can be used
```

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_lower(sns.kdeplot)
g.map_upper(sns.swarmplot)
```

# Regression Plots

```python
import seaborn as sns
tips= sns.load_dataset('tips')
tips.head(3)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'])
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex',markers=['o','v'],scatter_kws={'s':100}) # s stands for size
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time') # works same a hue but it seperates it by row and co
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',hue='day')
```
