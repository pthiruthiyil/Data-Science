```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([.1,.1,.8,.8])
axes.set_xlabel("X Label")
axes.set_ylabel("Y Label")
axes.set_title("Title")
axes.plot(x,y)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes1 = fig.add_axes([0.2,0.3,0.3,0.5])
axes2 = fig.add_axes([0.55,0.3,0.3,0.5])

axes1.plot(x,y,'r')
axes2.plot(y,x,'b')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2) #tuple output
axes[0].plot(x,y)
axes[1].plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=3)
fig.tight_layout()
```

```python
fig = plt.figure(figsize=(3,3),dpi=100) # The same argument can be added to subplots methond as well 
fig.add_axes([0,0,.8,0.8])
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label="X Squared")
axes.plot(x,x**3,label="X Cubed")
axes.legend(loc=0) # we can provide a tuble loc=(0.1,0.1)
```
